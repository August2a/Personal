```python
n, m = list(map(int,input().split()))
qt = int(input())

matriz = []
for i in range(n):
    lista = []
    for j in range(m):
        lista.append(0)
    matriz.append(lista)
```

```python
maior_buraco = [0,0]

for _ in range(qt):
    x, y = list(map(int,input().split()))

    for i in range (n):
        for j in range(m):
            if i == x-1 or j == y-1:
                matriz[i][j] += 1

                if matriz[i][j] > maior_buraco[0]:
                    maior_buraco = [matriz[i][j],1]
                elif matriz[i][j] == maior_buraco[0]:
                    maior_buraco[1] += 1

print(maior_buraco[1])

```
